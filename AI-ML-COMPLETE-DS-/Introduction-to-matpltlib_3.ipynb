```python
car_sales=pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].astype(str).str.replace(r'[^0-9.]', '', regex=True).astype(float)
car_sales
```

```python
car_sales["Price"].dtype
```

```python
car_sales["Price"]=car_sales["Price"].astype(int)
```

```python
car_sales
```

```python
car_sales["Price"].dtype
```

```python
# if price dtype is af string then for removing last 2 zeros type .str[:2]
# if price dtype is of int then for removing last 2 zeros type price/100
```

```python
car_sales["Sales_Date"]=pd.date_range("5/2/24", periods=len(car_sales))
car_sales
```

```python
car_sales["Total_Sales"]=car_sales["Price"].cumsum()
```

```python
car_sales
```

```python
car_sales.plot(x="Sales_Date",y="Total_Sales");
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```
