## Customizing the plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# see the different styles available
plt.style.available
```

```python
car_sales=pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].astype(str).str.replace(r'[^0-9.]', '', regex=True).astype(float)
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].astype(int)
```

```python
car_sales
```

```python
heart_disease=pd.read_csv("heart-disease.csv")
```

```python
over_age_40=heart_disease[heart_disease["age"]>40]
```

```python
# Subplot of chol, age, thalach
fig, (ax1, ax2)= plt.subplots(nrows=2, ncols=1, figsize=(4,5), sharex=True);
plt.style.use("ggplot")
# Add data to ax0
k=ax1.scatter(x=over_age_40["age"],
           y=over_age_40["chol"],
           c=over_age_40["target"],
             cmap="winter");

# Customize ax1
ax1.set(title="Heart Disease and Cholesterol Levels",
       ylabel="cholesterol");

# Add a legend to ax1
ax1.legend(*k.legend_elements(), title="target");

# Add a meanline
ax1.axhline(over_age_40["age"].mean(), linestyle="--");

# for ax2
j=ax2.scatter(x=over_age_40["age"],
             y=over_age_40["thalach"],
             c=over_age_40["target"],
             cmap="plasma");
ax2.set(title="Heart Disease and Thalach Levels",
        xlabel="age",
       ylabel="thalach");

# change ax2 x,y axis limit
ax2.set_xlim([50,80])
ax2.set_ylim([100,180])
ax2.legend(*j.legend_elements(), title="target");
ax2.axhline(over_age_40["thalach"].mean(), linestyle="--");

# Add a title to the figure
fig.suptitle("heart disease analysis", fontsize=17, fontweight="bold");
```
