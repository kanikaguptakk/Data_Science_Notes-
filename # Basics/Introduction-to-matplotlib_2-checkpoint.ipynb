## Making figures using NumPy arrays
We want:
* Line Plot (default matplotlib plot is a line plot)
* Scatter Plot
* Histogram (numerical distribution)
* Bar Plot
* Sub Plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# plot the data and create a line plot 
x=np.linspace(0.2,100,10)
fig,ax=plt.subplots()
ax.plot(x,x**2);
```

```python
# use the same data to make a scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# another scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# one more scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.log(x));
```

```python
# matplotlib is versatile and flexible
```

```python
# make a plot from the dictionary
eatable_items={"toffees": 12," chocolates": 22, "lemon": 33, "ice-creams": 44}
fig,ax=plt.subplots()
ax.bar(eatable_items.keys(), height=eatable_items.values());
ax.set(title="Eatables", xlabel="eatable_names", ylabel="counts");
```

```python
# to make a plot horizontal in view
fig,ax=plt.subplots()
ax.barh(list(eatable_items.keys()), list(eatable_items.values()));
```

```python
# make some data for histograms and plot it
x=np.random.randn(100)
fig,ax=plt.subplots()
ax.hist(x);
```

```python
# subplots option 1
# create multiple subsets
fig,((ax1, ax2), (ax3, ax4))= plt.subplots(nrows=2, ncols=2, figsize=(3,5))
# Plot data to each axis
ax1.plot(x,x**4);
ax2.scatter(np.random.random(10), np. random.random(10));
ax3.bar(eatable_items.keys(), eatable_items.values());
ax4.hist(np.random.randn(210));
```

```python
# subplots option 2
fig, ax=plt.subplots(nrows=2, ncols=2, figsize=(3,5))
# plot to each diffrent index
ax[0,0].plot(x,x**4);
ax[0,1].scatter(np.random.random(100), np.random.random(100));
ax[1,0].bar(eatable_items.keys(), eatable_items.values());
ax[1,1].hist(np.random.randn(100));
```
