#### Matplotlib Example Workflow

```python
# import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# prepare the data
a=[1,3,4,6,7]
b=[5,6,7,8,9]

# setup plot
fig,ax=plt.subplots(figsize=(4,6))  # (width, height)

# plot the data
ax.plot(a,b)

# customize plot
ax.set(title="My first Plot", xlabel="x-axis", ylabel="y-axis" )

# save and shoe (you save the whole figure)
fig.savefig("")
```
